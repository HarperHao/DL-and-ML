```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#读取数据
train=pd.read_csv(r"K:\编程\机器学习数据资料\交通事故理赔审核\train.csv")
test=pd.read_csv(r"K:\编程\机器学习数据资料\交通事故理赔审核\test.csv")
submit=pd.read_csv(r"K:\编程\机器学习数据资料\交通事故理赔审核\sample_submit.csv")
```

```python
train.info()
```

```python
train.describe() 
```

```python
#删除ID这一列
train.drop("CaseId",axis=1,inplace=True)
test.drop("CaseId",axis=1,inplace=True)
```

```python
#不同变量间的相关系数
train.corr()
```

```python
#对相关系数矩阵进行处理
corr=train.corr()
#负相关不代表没关系
corr[corr<0]=np.nan
```

```python
corr
```

```python
corr[np.abs(corr)<0.1]=np.nan
corr
```

```python
y_train=train.pop('Evaluation')
```

```python
#Logistic回归模型,加了正则项
clf=LogisticRegression(penalty='l1',C=1.0,random_state=0)
clf.fit(train,y_train)
y_pred=clf.predict_proba(test)[:,1]
```

```python
submit['Evaluation']=y_pred
submit.to_csv(r'K:\编程\机器学习数据资料\交通事故理赔审核\my_LASSO_prediction.csv', index=False)
```
