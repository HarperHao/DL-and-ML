```python
import numpy as np
import scipy as sp
from scipy.optimize import leastsq
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#目标函数
def real_func(x):
    return np.sin(2*np.pi*x)
#多项式（拟合函数）
def fit_func(p,x):
    # np.poly1d([1,2,3])->1x^2+2x^1+3x^0
    f=np.poly1d(p)
    #输出拟合函数
    #print('拟合函数为：',f)
    #返回具体的值
    return f(x)
#计算残差
def residuals_func(p,x,y):
    ret=fit_func(p,x)-y
    return ret**2
```

```python
#已知十个点来拟合曲线
x=np.linspace(0,1,10)
x_points=np.linspace(0,1,1000)
#y_是目标函数的值
y_=real_func(x)
#加上正态分布噪音的目标函数的值
y=[np.random.normal(0,0.1)+y1 for y1 in y_]
```

```python
def fitting(M=0):
    """
    M为多项式的次数
    """
    #随机初始化多项式参数
    #p_init是一个M+1维的（0，1）的均匀分布样本值,确认初始有几项
    p_init=np.random.rand(M+1)
    #最小二乘法
    p_lsq=leastsq(residuals_func,p_init,args=(x,y))
    print("Fitting Parameters:",p_lsq[0])
 
    #可视化
    #真实值
    plt.plot(x_points,real_func(x_points),label='real')
    #拟合曲线
    plt.plot(x_points,fit_func(p_lsq[0],x_points),label='fitted curve')
    #噪点
    plt.plot(x,y,'bo',label='noise')
    plt.legend()
    return p_lsq
```

```python

p_lsq_0=fitting(M=1)
```

一些函数的学习：
1. scipy库中的leastsq函数，可以求出拟合曲线的参数。
2. np中的poly1d函数，用来构造拟合曲线方程。

思路：
1. 3个函数，分别用来求取真实函数的值，拟合曲线的值和残差。
2. 构造10个样本点进行拟合曲线参数的训练，此处有两步，一步是构造是个样本点，要用到linspace函数,最优参数由leastsq函数得出。
3. 可视化处理，用1000个点，画出真实曲线，噪点，拟合曲线。
4. 2、3步要放到fitting函数中。


```python
#正则化处理
regularization=0.0001
def residuals_func_regularization(p,x,y):
    ret=fit_func(p,x)-y
    ret=np.append(ret,np.sqrt(0.5*regularization*np.square(p)))
 
    return ret**2
```

```python
#最小二乘法，加正则化项
p_init=np.random.rand(9+1)
p_lsq_regularization=leastsq(residuals_func_regularization,p_init,args=(x,y))
```

```python

#可视化
#真实值
plt.plot(x_points,real_func(x_points),label='real')
#未正则化的拟合曲线
plt.plot(x_points,fit_func(p_lsq_9[0],x_points),label='fitted curve')
#正则化的拟合曲线
plt.plot(x_points,
        fit_func(p_lsq_regularization[0],x_points),
        label='regularization')
#噪点
plt.plot(x,y,'bo',label='noise')
plt.legend()
```
